```python
X = df.drop(columns=['BurnoutRisk'])
y= df['BurnoutRisk']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
#Encoding categorical variables
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
transformer = ColumnTransformer(transformers=[('tnf1',OrdinalEncoder(),['HasMentalHealthSupport','HasTherapyAccess']),
                                              ('tnf2',OneHotEncoder(),['Gender','Country','JobRole','Department','RemoteWork','SalaryRange'])],
                                              remainder = 'passthrough')
```

```python
X_train = transformer.fit_transform(X_train)
```

```python
X_test = transformer.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
